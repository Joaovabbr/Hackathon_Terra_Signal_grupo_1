```python
%sql

CREATE SCHEMA IF NOT EXISTS workspace.churn_zero;
```

```python
file_path = "/Workspace/Users/joaovab@al.insper.edu.br/Hackathon_Terra_Signal_grupo_1/history.csv"

df_raw = spark.read.csv(file_path, header=True, inferSchema=True)

new_columns = [col_name.replace(" ", "_").replace("(", "").replace(")", "") for col_name in df_raw.columns]
df_raw = df_raw.toDF(*new_columns)

target_table = "workspace.churn_zero.historico_bronze" 

print(f"Salvando dados em {target_table}...")

df_raw.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable(target_table)

print("Sucesso! Tabela criada.")
display(spark.sql(f"SELECT * FROM {target_table} LIMIT 5"))
```
